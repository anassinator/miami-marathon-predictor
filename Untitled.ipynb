```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df  = pd.read_csv('data.csv')
```

```python
df = df[df['ignore'] == False]
```

```python
df['time_in_secs'] = df['time'].apply(lambda x: x.split(':')).apply(lambda x:int(x[0])* 360 + int(x[1])* 6 0 + int(x[2]))
```

```python
df
```

```python
sns.jointplot(x='age',y='time_in_secs',data=df,kind = 'kde')
```

```python
sns.jointplot(x='age',y='time_in_secs',data=df,kind = 'reg')
```

```python
sns.pairplot(df,hue = 'male', kind = 'reg')
```

```python
sns.pairplot(df,kind = 'reg')
```
